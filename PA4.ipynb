```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('board2.csv')
df
```

#### Problem 2

```python
df['Average'] = df[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)

# Visualization for Track
plt.figure(figsize=(14, 6))
sns.barplot(x='Track', y='Average', data=df)
plt.title('Average Score by Track')
plt.xlabel('Track')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()

# Visualization for Gender
plt.figure(figsize=(10, 6))
sns.barplot(x='Gender', y='Average', data=df)
plt.title('Average Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.show()

# Visualization for Hometown
plt.figure(figsize=(14, 6))
sns.barplot(x='Hometown', y='Average', data=df)
plt.title('Average Score by Hometown')
plt.xlabel('Hometown')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.show()
```
